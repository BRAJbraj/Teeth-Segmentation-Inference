```python
# Function to download files from Google Drive using gdown
# Make sure the files are publicly accessible ("Anyone with the link")
def download_from_drive(file_id, output_path):
    print(f"Downloading {output_path}...")
    # 'gdown' is a command-line utility preinstalled in Colab environments
    os.system(f'gdown --id {file_id} -O {output_path}')
    print("Download complete.")

# Function to unzip files to a specified target directory
def unzip_file(zip_name, extract_path):
    print(f"Extracting {zip_name} to {extract_path}...")
    os.makedirs(extract_path, exist_ok=True) # Ensure the destination directory exists
    with ZipFile(zip_name, 'r') as zip_ref:
        zip_ref.extractall(extract_path) # Extract all contents to the specified path
    print("Extraction complete.")

# Function to merge contents of a source folder into a destination folder
def merge_folders(source_folder, destination_folder):
    print(f"Merging contents of {source_folder} into {destination_folder}...")
    for item in os.listdir(source_folder):
        s = os.path.join(source_folder, item)
        d = os.path.join(destination_folder, item)
        if os.path.isdir(s):
            # Recursively copy directories, allowing existing dirs
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d) # Copy single files
    print("Merge complete.")
```

```python
import pandas as pd
import os
import shutil

def organize_files_for_inference(csv_path, source_base_dir, destination_base_dir):
    """
    Reads a CSV of file paths, copies those files from the source directory,
    and places them in a destination directory while preserving the folder structure.

    Args:
        csv_path (str): Path to the CSV file containing 'file_path' column.
        source_base_dir (str): The root directory where original files are located.
        destination_base_dir (str): The root directory where files will be copied to.
    """
    print(f"Starting file organization process...")

    # Ensure the destination base directory exists
    if not os.path.exists(destination_base_dir):
        os.makedirs(destination_base_dir)
        print(f"Created destination directory: {destination_base_dir}")

    # Read the CSV file
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f"Error: CSV file not found at {csv_path}")
        return
    except Exception as e:
        print(f"An error occurred while reading CSV: {e}")
        return

    # Assuming the CSV has a column named 'file_path' with full paths
    if 'file_path' not in df.columns:
        print("Error: CSV must contain a 'file_path' column.")
        return

    count = 0
    for full_path in df['file_path']:
        # Extract the relative path from the source base directory
        # e.g., turn '/content/ToothGroupNetwork/data_obj_parent_directory/013FHA7K/file.obj'
        # into '013FHA7K/file.obj'
        relative_path = os.path.relpath(full_path, source_base_dir)

        # Define the new destination path
        destination_path = os.path.join(destination_base_dir, relative_path)

        # Ensure the destination directory for the specific file exists
        destination_dir = os.path.dirname(destination_path)
        if not os.path.exists(destination_dir):
            os.makedirs(destination_dir)

        # Copy the file
        try:
            shutil.copy2(full_path, destination_path)
            # print(f"Copied: {relative_path}") # Uncomment for verbose logging
            count += 1
        except FileNotFoundError:
            print(f"Warning: Source file not found: {full_path}")
        except Exception as e:
            print(f"An error occurred while copying {full_path}: {e}")

    print(f"\nFinished copying {count} files.")
    print(f"Files are organized in: {destination_base_dir}")


if __name__ == "__main__":
    # --- Configuration ---
    # !! UPDATE these paths to match your environment !!
    CSV_FILE = "/content/patient_paths.csv"
    SOURCE_ROOT = "/content/ToothGroupNetwork/data_obj_parent_directory/"
    DESTINATION_ROOT = "/content/inference_ready_data/"
    # ---------------------

    organize_files_for_inference(CSV_FILE, SOURCE_ROOT, DESTINATION_ROOT)

```

```python
# ******************************************************************************************************
#  Code for extracting full path for files needed for inferencing from /content/base_name_test_fold.txt
# ******************************************************************************************************

# Colab-ready script
# Purpose: for each patient id listed in base_name_test_fold.txt, print full paths
# of JSON and OBJ files found under the provided parent directories.
#
# Edit the three variables below if needed before running.

import os
import csv
import fnmatch
import re

# ---------- EDIT THESE PATHS ----------
json_parent_dir = "/content/ToothGroupNetwork/data_json_parent_directory"
obj_parent_dir  = "/content/ToothGroupNetwork/data_obj_parent_directory"
patient_list_txt = "/content/base_name_test_fold.txt"
# Output CSV (set to None to skip writing)
out_csv = "/content/patient_paths.csv"
# --------------------------------------

# Keywords to detect 'upper' / 'lower' in filenames or folder names
UP_KEYWORDS = ["upper", "_U", "_u", "upp", "maxilla", "maxillary"]
LOW_KEYWORDS = ["lower", "_L", "_l", "low", "mand", "mandible", "mandibular"]

def contains_keyword(name, keywords):
    name_l = name.lower()
    for kw in keywords:
        if kw.lower() in name_l:
            return True
    return False

def find_files_under_dir(base_dir, patient_id, extensions=(".json", ".obj")):
    """
    Search recursively for files under base_dir that are inside a patient-specific folder
    or whose filename contains the patient_id. Returns a list of absolute paths.
    """
    found = []
    pt_dir = os.path.join(base_dir, patient_id)
    if os.path.isdir(pt_dir):
        # walk only inside that patient directory (fast and accurate)
        for root, _, files in os.walk(pt_dir):
            for fn in files:
                if fn.lower().endswith(tuple(e.lower() for e in extensions)):
                    found.append(os.path.join(root, fn))
        return found
    # fallback: search entire base_dir for files that include the patient_id in the filename
    for root, _, files in os.walk(base_dir):
        for fn in files:
            if fn.lower().endswith(tuple(e.lower() for e in extensions)) and patient_id.lower() in fn.lower():
                found.append(os.path.join(root, fn))
    return found

def classify_by_side(paths):
    """
    Given a list of file paths, try to classify each as 'upper', 'lower' or 'unknown'
    based on filename or immediate parent folder name.
    Returns dict: {'upper': [...], 'lower': [...], 'unknown': [...]}
    """
    groups = {"upper": [], "lower": [], "unknown": []}
    for p in paths:
        fn = os.path.basename(p)
        parent = os.path.basename(os.path.dirname(p))
        if contains_keyword(fn, UP_KEYWORDS) or contains_keyword(parent, UP_KEYWORDS):
            groups["upper"].append(p)
        elif contains_keyword(fn, LOW_KEYWORDS) or contains_keyword(parent, LOW_KEYWORDS):
            groups["lower"].append(p)
        else:
            groups["unknown"].append(p)
    return groups

def load_patient_ids(txt_path):
    with open(txt_path, "r") as f:
        lines = [ln.strip() for ln in f.readlines()]
    # filter out blank lines and comments
    ids = [ln for ln in lines if ln and not ln.startswith("#")]
    return ids

# ------------------------------
# Main logic
# ------------------------------
patient_ids = load_patient_ids(patient_list_txt)
print(f"Found {len(patient_ids)} patient ids in: {patient_list_txt}\n")

results = []  # list of dicts to write to CSV if requested

for pid in patient_ids:
    info = {"patient_id": pid,
            "json_upper": [], "json_lower": [], "json_unknown": [],
            "obj_upper": [],  "obj_lower": [],  "obj_unknown": []}
    # find JSONs
    json_paths = find_files_under_dir(json_parent_dir, pid, extensions=(".json",))
    json_groups = classify_by_side(json_paths)
    info["json_upper"] = json_groups["upper"]
    info["json_lower"] = json_groups["lower"]
    info["json_unknown"] = json_groups["unknown"]

    # find OBJs
    obj_paths = find_files_under_dir(obj_parent_dir, pid, extensions=(".obj",))
    obj_groups = classify_by_side(obj_paths)
    info["obj_upper"] = obj_groups["upper"]
    info["obj_lower"] = obj_groups["lower"]
    info["obj_unknown"] = obj_groups["unknown"]

    # Print a concise summary for this patient
    print(f"Patient: {pid}")
    def p_list(name, lst):
        print(f"  {name}: {len(lst)}")
        # print up to 10 items; if none, show '-'
        if lst:
            for p in lst[:10]:
                print(f"    - {p}")
            if len(lst) > 10:
                print(f"    ... ({len(lst)-10} more)")
        else:
            print("    - (none)")

    p_list(" JSON upper", info["json_upper"])
    p_list(" JSON lower", info["json_lower"])
    p_list(" JSON unknown", info["json_unknown"])
    p_list(" OBJ upper", info["obj_upper"])
    p_list(" OBJ lower", info["obj_lower"])
    p_list(" OBJ unknown", info["obj_unknown"])
    print("")

    results.append(info)

# ------------------------------
# Optionally write CSV with one row per patient + joined paths
# ------------------------------
if out_csv:
    fieldnames = [
        "patient_id",
        "json_upper", "json_lower", "json_unknown",
        "obj_upper", "obj_lower", "obj_unknown"
    ]
    with open(out_csv, "w", newline="") as csvf:
        writer = csv.DictWriter(csvf, fieldnames=fieldnames)
        writer.writeheader()
        for row in results:
            writer.writerow({
                "patient_id": row["patient_id"],
                "json_upper": ";".join(row["json_upper"]),
                "json_lower": ";".join(row["json_lower"]),
                "json_unknown": ";".join(row["json_unknown"]),
                "obj_upper": ";".join(row["obj_upper"]),
                "obj_lower": ";".join(row["obj_lower"]),
                "obj_unknown": ";".join(row["obj_unknown"]),
            })
    print(f"Saved summary CSV to: {out_csv}")
else:
    print("CSV writing disabled (out_csv is None).")

# results is available as a Python list of dicts for further processing.

```

```python
#**********************************************************************************
# Read the csv file and print the full path by proper grouping
# *********************************************************************************


import csv
from collections import OrderedDict
from itertools import product

# ---------- CONFIG ----------
CSV_PATH = "/content/patient_paths.csv"   # adjust if needed
PAIRING_MODE = "first"   # "first" or "all"
SAVE_PAIRS_TXT = None    # e.g. "/content/inference_pairs.txt" or None to skip
# ----------------------------

def split_paths(cell):
    """Split semicolon-separated paths; return list (empty if cell blank)."""
    if cell is None:
        return []
    cell = cell.strip()
    if not cell:
        return []
    # Some entries may be quoted; strip quotes
    if (cell.startswith('"') and cell.endswith('"')) or (cell.startswith("'") and cell.endswith("'")):
        cell = cell[1:-1]
    parts = [p.strip() for p in cell.split(";") if p.strip()]
    return parts

def load_grouped_csv(csv_path):
    grouped = OrderedDict()  # keep order of patients as in CSV
    with open(csv_path, newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            pid = row.get("patient_id") or row.get("patient") or row.get("id")
            if pid is None:
                continue
            entry = {
                "json_upper": split_paths(row.get("json_upper","")),
                "json_lower": split_paths(row.get("json_lower","")),
                "json_unknown": split_paths(row.get("json_unknown","")),
                "obj_upper": split_paths(row.get("obj_upper","")),
                "obj_lower": split_paths(row.get("obj_lower","")),
                "obj_unknown": split_paths(row.get("obj_unknown","")),
            }
            grouped[pid] = entry
    return grouped

def print_grouped_for_manual_inference(grouped):
    for pid, e in grouped.items():
        print(f"Patient: {pid}")
        for side in ("upper", "lower"):
            jsons = e.get(f"json_{side}", [])
            objs  = e.get(f"obj_{side}", [])
            print(f"  {side.upper()}:")
            print(f"    JSONs: {len(jsons)}")
            for p in jsons[:10]:
                print(f"      - {p}")
            if len(jsons) > 10:
                print(f"      ... ({len(jsons)-10} more)")
            print(f"    OBJs:  {len(objs)}")
            for p in objs[:10]:
                print(f"      - {p}")
            if len(objs) > 10:
                print(f"      ... ({len(objs)-10} more)")
        # unknown (if any)
        if e.get("json_unknown") or e.get("obj_unknown"):
            print("  UNKNOWN / UNCATEGORIZED:")
            for p in e.get("json_unknown", [])[:10]:
                print(f"    JSON - {p}")
            for p in e.get("obj_unknown", [])[:10]:
                print(f"    OBJ  - {p}")
        print("")  # blank line between patients

def generate_pairs_for_inference(grouped, mode="first", include_incomplete=False):
    """
    Returns a list of tuples (patient_id, side, json_path or None, obj_path or None).
    mode:
      - "first": pick first JSON and first OBJ (if present) for each patient+side
      - "all": produce all combinations json x obj for that patient+side
    include_incomplete:
      - if True, include pairs where one side is None (json or obj missing)
      - if False, skip incomplete pairs
    """
    pairs = []
    for pid, e in grouped.items():
        for side in ("upper", "lower"):
            jsons = e.get(f"json_{side}", [])
            objs  = e.get(f"obj_{side}", [])
            if mode == "first":
                j = jsons[0] if jsons else None
                o = objs[0] if objs else None
                if j is None and o is None:
                    continue
                if (j is None or o is None) and not include_incomplete:
                    continue
                pairs.append((pid, side, j, o))
            elif mode == "all":
                if jsons and objs:
                    for j_path, o_path in product(jsons, objs):
                        pairs.append((pid, side, j_path, o_path))
                else:
                    # if one list empty -> include only if include_incomplete True
                    if include_incomplete:
                        if jsons:
                            for j_path in jsons:
                                pairs.append((pid, side, j_path, None))
                        if objs:
                            for o_path in objs:
                                pairs.append((pid, side, None, o_path))
                    # else skip
            else:
                raise ValueError("mode must be 'first' or 'all'")
    return pairs

# --------------------------
# Run
# --------------------------
grouped = load_grouped_csv(CSV_PATH)
print(f"Loaded {len(grouped)} patients from {CSV_PATH}\n")

# Print grouped content for manual inspection / copying
print_grouped_for_manual_inference(grouped)

# Generate inference pairs according to chosen mode
pairs = generate_pairs_for_inference(grouped, mode=PAIRING_MODE, include_incomplete=False)
print(f"Generated {len(pairs)} inference pairs with mode='{PAIRING_MODE}' (incomplete pairs skipped)\n")

# Show first 50 pairs as convenient copy/paste lines
for i, (pid, side, json_p, obj_p) in enumerate(pairs[:50]):
    json_display = json_p if json_p is not None else "<MISSING_JSON>"
    obj_display  = obj_p  if obj_p  is not None else "<MISSING_OBJ>"
    print(f"{pid}\t{side}\t{json_display}\t{obj_display}")

# Optionally save pairs to a simple TSV for downstream scripts
if SAVE_PAIRS_TXT:
    with open(SAVE_PAIRS_TXT, "w", newline="") as out:
        out.write("patient_id\tjaw\tjson_path\tobj_path\n")
        for pid, side, json_p, obj_p in pairs:
            out.write(f"{pid}\t{side}\t{json_p or ''}\t{obj_p or ''}\n")
    print(f"\nSaved inference pairs to: {SAVE_PAIRS_TXT}")
```
